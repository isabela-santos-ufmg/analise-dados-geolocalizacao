# Complete map with all data collect
We'll start mixing the to .csv's, trasnforming in to a DataFrame and use the .concat() function

```python
import pandas as pd
import folium
from folium.plugins import Geocoder # para barra de pesquisa
from branca.element import Element # para legenda
```

```python
df1 = pd.read_csv('geolocated_bars.csv')
df2 = pd.read_csv('geolocated_places_complete.csv')

sub1 = pd.DataFrame()
sub2 = pd.DataFrame()
```

```python
df1
```

```python
df2
```

```python
subcolumns = ['name', 'category', 'geolocated_address', 'ref', 'website', 'latitude', 'longitude']
sub1 = df1[subcolumns]
sub2 = df2[subcolumns]

df = pd.concat([sub1, sub2])
df
```

```python
# standardizing names
df['name'] = df['name'].str.upper()

# correcting names
df.loc[df['name'] == 'FINALMENTE', 'name'] = 'FINALMENTE CLUBE'
df.loc[df['name'] == 'CONSTRUCTION BAR - BAIRRO ALTO', 'name'] = 'CONSTRUCTION'
df.loc[df['name'] == 'SHELTER BAR', 'name'] = 'SHELTER'

#cleaning addresses
df = df.dropna(subset=['geolocated_address', 'latitude', 'longitude'])
df = df[df['name'] != 'WOOFLX (FECHADO)	'].copy()
df = df[~df['name'].isin(['WOOFLX (FECHADO)'])].copy()
df = df[df['name'] != '106'].copy()
df = df[~df['name'].isin(['106'])].copy()


#removing duplicates
df_limpo = df.drop_duplicates(subset=['name'])

# saving
df_limpo.to_csv('completed_places.csv', index=False, encoding='utf-8-sig')
df_limpo
```

## Versão melhorada com sites e localizaçao pontual 

Vamos obter a versão final da csv com os sites e localização pontual dos estabelecimentes


```python
# pegando complete_places.csv e editando a mão para completar valores
df_final = pd.read_csv("completed_places_edited.csv")
```

```python
import geopy.geocoders
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
import pandas as pd
import ssl
import certifi
```

```python
ctx = ssl.create_default_context(cafile=certifi.where())
geopy.geocoders.options.default_ssl_context = ctx
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE

localizer = Nominatim(scheme = 'http', user_agent="my_geolocation_places", timeout=10)
geocode = RateLimiter(localizer.geocode, min_delay_seconds=1)
```

```python
locations = []
for i, row in df_final.iterrows():
    name = row['name']
    address = row['geolocated_address']
    location = geocode(address)
    locations.append(location)
    if location:
        print(f"-> found!: {name}")
    else:
        location = None
        print(f"-> NOT found.{name}")

print("\ngeocoding finished.")
```

```python
df_final['completed_address'] = [loc.address if loc else None for loc in locations]
df_final['latitude'] = [loc.latitude if loc else None for loc in locations]
df_final['longitude'] = [loc.longitude if loc else None for loc in locations]
df_final['point'] = [tuple(loc.point) if loc else None for loc in locations]

#cleaning
df_final['name'] = df_final['name'].str.upper()
df_final['name'] = df_final['name'].str.strip()
df_final['category'] = df_final['category'].astype(str).str.strip().str.lower()
df_final
```

```python
df_final.to_csv('final_places.csv', index=False, encoding='utf-8-sig')
```

# Plotando em um mapa usando Folium

https://getbootstrap.com/docs/3.3/components/

https://fontawesome.com/v4/icons/

```python
categories_config = {
    'nightclub/bar': {'color': 'pink', 'icon':'glyphicon glyphicon-glass' , 'prefix': 'glyphicon'},       
    'sauna': {'color':'darkred', 'icon':'fa-bath', 'prefix': 'fa'},              
    'restaurant': {'color':'orange', 'icon':'glyphicon glyphicon-cutlery' , 'prefix': 'glyphicon'},        
    'association': {'color':'purple', 'icon':'glyphicon glyphicon-heart-empty' , 'prefix': 'glyphicon'},      
    'accommodation': {'color':'green', 'icon':'fa-bed', 'prefix': 'fa'},    
    'cultural space': {'color':'gray', 'icon':'film' , 'prefix': 'glyphicon'},  
    'store': {'color':'lightblue', 'icon':'glyphicon glyphicon-shopping-cart' , 'prefix': 'glyphicon'}, #or glyphicon glyphicon-tag
    }
```

```python
subtitle_html = """
<div style="
    position: fixed;
    bottom: 20px;
    right: 20px;
    z-index: 1000;
    background: white;
    padding: 10px;
    border: 1px solid gray;
    opacity: 0.9;
    font-family: Arial;
    max-height: 200px;
    overflow-y: auto;
">
    <b>Subtitle</b>
"""
for name, config in categories_config.items():
    subtitle_html += f"""
    <div style="margin: 5px 0;">
        <span style="color: {config['color']}; font-size: 18px;">■</span>
        {name}
    </div>
    """

subtitle_html += "</div>"

subtitle = Element(subtitle_html)
```

```python
default_color = 'black'
center = [38.7223, -9.1393]
lisbon_map = folium.Map(location=center, zoom_start=14)
lisbon_map.get_root().html.add_child(subtitle)

Geocoder().add_to(lisbon_map)

for i, row in df_final.iterrows():
    if pd.notna(row['latitude']) and pd.notna(row['longitude']):
        coordinates = [row['latitude'], row['longitude']]
        name = row['name']
        address = row['completed_address']
        category = row['category']
        site = row['website']
            
        icon_config = categories_config.get(category, {
                                             'color': default_color,
                                             'icon': 'glyphicon glyphicon-remove',
                                             'prefix': 'glyphicon'
                                            })
        html_popup = f"""
        <h4>{name}</h4>
        <p>
        <b>Address: </b> {address}<br>
        </p>
        <p>
        <b>Category: </b> {category}<br>
        </p>
        <p>
        <b>Site: </b> {site}<br>
        </p>
        """
        popup = folium.Popup(html_popup, max_width=250)
        folium.Marker(
            location=coordinates,
            popup=popup, 
            tooltip=name, 
            icon=folium.Icon(
                icon = icon_config['icon'],
                color= icon_config['color'],
                prefix= icon_config['prefix'],
                icon_color='white'
                ) ).add_to(lisbon_map)

lisbon_map.save('lisbon_places_map.html')
lisbon_map
```
